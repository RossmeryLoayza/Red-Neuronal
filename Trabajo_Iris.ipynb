```
# from sklearn.datasets import load_iris

# iris = load_iris()
# iris

import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers.core import Dense

from google.colab import drive
drive.mount('/content/drive')

```

```

entrada = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Iris/entrada_iris.csv')
deseado = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Iris/deseado.csv')
entrada
```

```
model = Sequential()
model.add(Dense(60,input_dim=4 , activation='relu'))
model.add(Dense(1, activation='sigmoid')) 
model.compile(loss='mean_squared_error', optimizer='adam', metrics='binary_accuracy')
model.fit(entrada,deseado,epochs=1000)
```

```
resultados = model.evaluate(entrada,deseado)
print("\n%s: %.2f%%" % (model.metrics_names[1],resultados[1]*100))
```
